```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('insurance.csv')
#smoker
df.replace(to_replace = 'male',
          value = 1,
          inplace = True)
df.replace(to_replace = 'female',
          value = 0,
          inplace = True)
#smoker
df.replace(to_replace = 'yes',
          value = 1,
          inplace = True)
df.replace(to_replace = 'no',
          value = 0,
          inplace = True)
# region
df.replace(to_replace = 'northeast',
          value = 0,
          inplace = True)
df.replace(to_replace = 'northwest',
          value = 1,
          inplace = True)
df.replace(to_replace = 'southeast',
          value = 2,
          inplace = True)
df.replace(to_replace = 'southwest',
          value = 3,
          inplace = True)
```

```python
df.head()
```

```python
sns.heatmap(df.corr(), annot=True)

plt.show()
```

```python
df.columns
```

```python
X = df[['age', 'sex', 'bmi', 'children', 'smoker', 'region']].to_numpy()
y = df['charges'].to_numpy()
m = X.size

mean = np.ones(X.shape[1])
std = np.ones(X.shape[1])
for i in range(0, X.shape[1]):
    mean[i] = np.mean(X.transpose()[i])
    std[i] = np.std(X.transpose()[i])
    for j in range(0, X.shape[0]):
        X[j][i] = (X[j][i] - mean[i])/std[i]
```

```python
plt.plot(X, y, 'b.')
plt.xlabel('$x_1$', fontsize=18)
plt.ylabel('$y$', rotation=0, fontsize=18)
plt.show()
```

```python
X_b = np.c_[np.ones((y.size,1)), X]
X_b
```

```python
theta_best = np.linalg.inv(X_b.T@X_b) @ X_b.T @ y
theta_best
```

```python
# X_new = np.array([[0], [1], [1], [1], [1], [1], [1]])
# X_new_b = np.c_[np.ones((7, 1)), X_new]
# X_new_b.shape

def predict_values(thetas, X_train):
    prediction = 0
    X_train = [1] + X_train
    for i in range(thetas.size):
        prediction += thetas[i] * X_train[i]
        
    return prediction
```

```python
print(predict_values(theta_best, [19, 1, 35.4, 0, 0, 3]))
print(predict_values(theta_best, [30, 1, 50.2, 2, 0, 0]))
print(predict_values(theta_best, [19, 0, 27.9, 0, 1, 3]))
```

```python
m = y.size
J = (1/(2*m))*((X-y.reshape(m, 1))**2).sum(axis=1) # MSE
J
```

```python

plt.xlabel(r'$\theta_1$', fontsize=12)
plt.ylabel(r'$J(\theta_1)$', fontsize=12)
plt.plot(y, J, 'b.', linewidth=2)
plt.grid(True)
plt.show()
```
